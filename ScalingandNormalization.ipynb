import the modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

importing data and the library for standardization

```python
df = pd.read_csv("/content/Iris.csv", usecols=["SepalLengthCm", "PetalLengthCm"])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
df_new = pd.DataFrame(sc.fit_transform(df), columns=df.columns) 
```

SCATTERPLOT BEFORE AND AFTER STANDARDIZATION

```python
plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
plt.title("Scatterplot Before Standardization", fontsize=18)
sns.scatterplot(data = df, color="blue")
plt.subplot(1,2,2)
plt.title("Scatterplot After Standardization", fontsize=18)
sns.scatterplot(data = df_new, color="red")
plt.tight_layout()
plt.show()
```

PDF before and after standardization

```python
plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
plt.title("PDF Before Standardization", fontsize=18)
sns.kdeplot(data = df, color="blue")
plt.subplot(1,2,2)
plt.title("PDF After Standardization", fontsize=18)
sns.kdeplot(data = df_new, color="red")
plt.tight_layout()
plt.show()
```

Normalization techniques -
MINMAX SCALING

```python
from sklearn.preprocessing import MinMaxScaler  #MinMaxScaler used for standardization
```

```python
mm = MinMaxScaler() 
df_new_mm = pd.DataFrame(mm.fit_transform(df), columns=df.columns) 
```

scatterplot of before and after Min Max Scaling

```python
plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
plt.title("Scatterplot Before Min Max Scaling", fontsize=18)
sns.scatterplot(data = df, color="blue")
plt.subplot(1,2,2)
plt.title("Scatterplot After Min Max Scaling", fontsize=18)
sns.scatterplot(data = df_new_mm, color="red")
plt.tight_layout()
plt.show()
```
