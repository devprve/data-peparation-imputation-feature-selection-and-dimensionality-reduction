````python
trainfile = '/content/netflix1.csv'
df = pd.read_csv(trainfile)
print(df.shape)
df.head()
````

before feature selection we have to split the dataset as train and test

````python
from sklearn.model_selection import train_test_split
target = 'type'
train, test = train_test_split(df, test_size=0.1, random_state=99,
                                stratify=df[target])
print(train.shape, test.shape)
````

````python
from autoviml.Auto_ViML import Auto_ViML
````

 feature selection using AutoViML

````python
model, features, trainm, testm = Auto_ViML(
    train,
    target,
    test,
    sample_submission='',
    hyper_param="RS",
    feature_reduction=True,
    scoring_parameter="weighted-f1",
    KMeans_Featurizer=False,
    Boosting_Flag="CatBoost",
    Binning_Flag=False,
    Add_Poly=True,
    Stacking_Flag=False,
    Imbalanced_Flag=False,
    verbose=1,
)
len(features)
````
